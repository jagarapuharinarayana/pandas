```python
import pandas as pd

# Sample DataFrame
data = {
    'Lab - 1': ['Mon 2 hours W101', 'Wed 2 hours W403', 'Mon 2 hours W106'],
    'Lab - 2': ['Thu 2 hours W102', 'Tue 2 hours W102', 'Thu 2 hours S412']
}

df = pd.DataFrame(data)

# Specify the custom sorting order for days
days_order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']

# Define a function to extract the day from a lab entry
def extract_day(entry):
    return entry.split()[0]

# Sort the Series based on the custom day order
sorted_series = df.loc[1].sort_values(key=lambda x: x.map(extract_day, na_action='ignore'))

# Print the sorted Series
print(sorted_series)
```
