```python
def computeAge(n):
    if n>0 and n <=20:
        return "child"
    elif n>20 and n<=40:
        return "young"
    else:
        return "old"
```

```python
df['ageCat'] = df['age'].apply(computeAge)
```

```python
def mapper():
    data = []
    for i,j in zip(df.iloc[:,-2],df.iloc[:,2]):
        if i == "child" and j == "male":
            data.append("boy")
        elif i == "child" and j == "female":
            data.append("girl")
        elif i == "young" and j == "female":
            data.append("yF")
        elif i == "young" and j == "male":
            data.append("yM")
        elif i == "old" and j == "female":
            data.append("women")
        elif i == "old" and j == "male":
            data.append("men")
    return data
        
df['newcatUpdate'] = mapper()
```
